```python
def recopilar( restricciones, columna):
    data = -1
    while not (data in restricciones):
        data = input(f"Ingrese {columna}: ")
        if data.upper() == 'X' and columna=="nivel":
            return "break"
        else:
            print("debe cumplir con la encuesta hasta el momento")
    return data

def informacion_recopilador():
    informacion = {"provincia":[],
                   "genero":[],
                   "nivel":[]}
    
    restricciones= [("L","A","P"),
                    ("M","F"),
                    ("1","2")]
    
    while True :
        restriccion = 0
        data = -1
        for columna in informacion.keys():
            data = recopilar(columna=columna, restricciones=restricciones[restriccion])
            if data=="break":
                return informacion 
            informacion[columna].append(data)
            restriccion+=1

def porcentajes(informacion):
    informacion = zip(informacion["provincia"], informacion["nivel"])
    total = len(informacion["nivel"])
    contadores_satisfaccion = [0,0,0]
    for provincia, nivel in informacion:
        if nivel == 1:
            match provincia:
                case "L": contadores_satisfaccion[0]+=1
                case "A": contadores_satisfaccion[1]+=1
                case "P": contadores_satisfaccion[2]+=1
    mensaje = f"""
    total 100
    lima x
    x =  lima*100 / total
    satisfaccion lima: {contadores_satisfaccion[0]*100/total}%
    satisfaccion arequipa: {contadores_satisfaccion[1]*100/total}%
    satisfaccion piura: {contadores_satisfaccion[2]*100/total}%
""" 
    return mensaje

def generos(informacion):
    informacion= zip(informacion["genero"],informacion["provincia"])
    contadores = [ [0,0] 
                  ,
                    [0,0] ,
                    [0,0] ]
    #h,m
    for genero, provincia in informacion:
        match provincia:
            case "L": 
                match genero:
                    case 'M': contadores[0][0]+=1
                    case 'F': contadores[0][1]+=1
            case "A": 
                match genero:
                    case 'M': contadores[1][0]+=1
                    case 'F': contadores[1][1]+=1
            case "P": 
                match genero:
                    case 'M': contadores[2][0]+=1
                    case 'F': contadores[2][1]+=1
    mensaje = f"""
    Lima:
        Masculino: {contadores[0][0]}
        Femenino: {contadores[0][1]}
    Arequipa:
        Masculino: {contadores[1][0]}
        Femenino: {contadores[1][1]}
    Piura:
        Masculino: {contadores[2][0]}
        Femenino: {contadores[2][1]}
    """     
    
informacion = informacion_recopilador()
porcentajes(informacion=informacion)
generos(informacion=informacion)
```

```python
def recopilar_texto( restricciones, columna):
    while not (data in restricciones):
        data = input(f"Ingrese {columna}: ")
    return data
def recopilar_rango(rangoa=0, rangob=10000, columna=""):
    data = -1
    while not(rangoa<data<rangob and type(data)==int):
        data = int(input(f"{columna}:"))
    return

def ICM(peso, altura):
    return peso/(altura**2)
def esta_en_riesgo(edad, asma, ICM):
    riesgo_uno = True if edad>60 else False
    riesgo_dos = True if asma=='S' else False
    riesgo_tres= True if ICM>=30 else False
    match (riesgo_uno, riesgo_dos, riesgo_tres).count(True):
        case 1:
            return "bajo"
        case 2: 
            return "medio"
        case 3:
            return "alto"
        case _:
            return "Sin riesgo"

def informacion_recopilador(numero_encuestados):
    informacion = []
    mensajes = ["Cual es su edad? ",
                 "Tiene asma?S/N ", 
                 "Cual es su altura(m)? ",
                 "cual es su peso?"]
    rangos = [(1,120), 
              ("SI", "NO"),
               (0.15, 2.30),
                (5, 200) ]
    while numero_encuestados>=0:
        infor_actual = []
        for indice, mensaje in enumerate(mensajes):
            if indice == 1:
                infor_actual.append(recopilar_texto(restricciones=rangos[indice], mensaje=mensaje))
            else:
                infor_actual.append(recopilar_rango(rangoa=rangos[indice][0], rangob=rangos[indice][1], columna=mensaje))
        infor_actual.append(ICM(peso=infor_actual[3], altura=infor_actual[2]))
        infor_actual.append(esta_en_riesgo(edad=infor_actual[1], asma=mensajes[2], ICM=infor_actual[4]))
        informacion.append(infor_actual)
        numero_encuestados-=1
        print(f"")
    return informacion

def estadisticas(informacion):
    contadores_riesgo = [0,0,0,0]
    for persona in informacion:
        match persona[5]:
            case 'bajo':
                contadores_riesgo[0]+=1
            case 'medio':
                contadores_riesgo[1]+=1
            case 'alto':
                contadores_riesgo[3]+=1
            case _: contadores_riesgo[4]+=1
    mensaje = f"""
    Riesgo bajo :{contadores_riesgo[0]}
    Riesgo medio :{contadores_riesgo[1]}
    Riesgo alt :{contadores_riesgo[2]}
    sin riesgo :{contadores_riesgo[3]}
    """
    return mensaje 
            
```
